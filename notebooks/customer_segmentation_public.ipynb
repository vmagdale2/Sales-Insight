```python
import pandas as pd
from sklearn.cluster import KMeans

# ✅ Step 1: Mount Google Drive (Do this manually in Colab before running)
from google.colab import drive
drive.mount('/content/drive')

# ✅ Step 2: Load Sales Data from Google Drive
file_path = "/redacted/sales_2_final.csv"
sales_df = load_data(file_path, usecols=["store_id", "product_id", "sales", "revenue"])

# ✅ Step 3: Check if Data Loaded Correctly
if sales_df.empty:
    print("❌ ERROR: Sales data did not load correctly. Check file path and format.")
else:
    print("✅ Data loaded successfully!")

    # ✅ Preprocess Data
    sales_df = preprocess_sales_data(sales_df)

    def preprocess_customer_data(df):
        """Creates customer features for segmentation."""
        customer_features = df.groupby("store_id").agg({
            "sales": "sum",
            "revenue": "sum",
            "product_id": "nunique"
        }).reset_index()

        customer_features.rename(columns={"product_id": "unique_products"}, inplace=True)
        return customer_features

    def run_kmeans(df, n_clusters=3):
        """Performs K-Means clustering on customer data."""
        model = KMeans(n_clusters=n_clusters, random_state=42)
        df["Cluster"] = model.fit_predict(df)
        return df

    # ✅ Step 4: Preprocess Customer Data
    customer_df = preprocess_customer_data(sales_df)

    # ✅ Step 5: Scale Features
    customer_df_scaled = scale_features(customer_df[["sales", "revenue", "unique_products"]])

    # ✅ Step 6: Run Clustering
    segmented_customers = run_kmeans(customer_df_scaled)

    # ✅ Step 7: Save Results to Google Drive
    output_path = "/redacted/processed_sales.csv"
    save_results(segmented_customers, output_path)

    print(f"✅ Customer Segmentation Completed! Results saved to: {output_path}")
```
